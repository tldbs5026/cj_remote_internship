### season

```python
#season

def season(x) :
    if x>=3 and x<= 5 :
        return 1
    elif x>=6 and x<= 8 :
        return 2
    elif x>=9 and x<=11 :
        return 3
    else :
        return 4
```

```python
da['season']=da['month'].apply(season)

da['season'].value_counts()
```
